```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
match=pd.read_csv('matches.csv')
delivery=pd.read_csv('deliveries.csv')
```

```python
# 1. Pie chart
# 2. Line chart
# 3. Scatter Plot 
# 4. Bar chart
# 5. Histogram
```

# PIE CHARTS

```python
x=delivery[delivery['batsman']=='V Kohli']
```

```python
y=x['batsman_runs'].value_counts()
```

```python
plt.pie(y, labels=y.index, autopct="%0.2f%%")
plt.show()
```

```python
def batsmanruns(batsman1):
    x=delivery[delivery['batsman']==batsman1]
    y=x['batsman_runs'].value_counts()
    plt.pie(y, labels=y.index, autopct="%0.2f%%")
    plt.show()
```

```python
batsmanruns('G Gambhir')
```

# LINE CHARTS

```python
#1.merging
```

```python
mega=match.merge(delivery,left_on='id',right_on='match_id')
```

```python
batsman_data=mega[mega['batsman']=='V Kohli']
```

```python
performance=batsman_data.groupby('season').sum()['batsman_runs']
```

```python
plt.plot(performance.index, performance.values, color="red",marker='D')
plt.xlabel('Season')
plt.ylabel('Runs scored')
plt.title("Batsman season by season performance")
plt.show()
```

```python
def batsmanper(batsman1,batsman2):
    mega=match.merge(delivery,left_on='id',right_on='match_id')
    batsman_data=mega[mega['batsman']==batsman1]
    batsman_data1=mega[mega['batsman']==batsman2]
    performance=batsman_data.groupby('season').sum()['batsman_runs']
    performance1=batsman_data1.groupby('season').sum()['batsman_runs']
    plt.plot(performance.index, performance.values,color="red",marker='D',label=batsman1)
    plt.plot(performance1.index, performance1.values,color="green",marker='+',label=batsman2)
    plt.xlabel('Season')
    plt.ylabel('Runs scored')
    plt.title("Batsman season by season performance")
    plt.legend()
    plt.show()
    
```

```python
batsmanper('MS Dhoni','AB de Villiers')
```
