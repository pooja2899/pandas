```python
import seaborn as sns
```

```python
tips=sns.load_dataset('tips')
```

```python
tips.head()
```

```python
# Distribution graphs
# -----> 1. Distplot
# -----> 2. Jointplot
# -----> 3. Pairplot
# -----> 4. Rugplot

#Categorical Plots
# -----> 1. Barplot
# -----> 2. Countplt
# -----> 3. Boxplot
# -----> 4. Violinplot

#Matrix Plot
# -----> 1. Heat map
```

# DISTRIBUTION GRAPHS

## 1. Distplot

```python
sns.distplot(tips['total_bill'])
```

```python
sns.distplot(tips['total_bill'], kde=False)
```

## 2.Jointplot

```python
sns.jointplot(x='total_bill',y='tip',data=tips)
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips, kind='hex')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips, kind='reg')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips, kind='kde')
```

## 3.Pairplot

```python
sns.pairplot(tips)
```

```python
sns.pairplot(tips, hue='sex')
```

## 4.Rugplot

```python
sns.rugplot(tips['total_bill'])
```

# CATEGORICAL PLOTS

## 1.Barplot

```python
sns.barplot(x='day',y='tip',data=tips)
```

```python
sns.barplot(x='day',y='tip',data=tips,hue='sex')
```

```python
sns.barplot(x='day',y='tip',data=tips,hue='smoker')
```

```python
import numpy as np
sns.barplot(x='day',y='tip',data=tips,hue='smoker',estimator=np.median)
```

## 2.Count Plot

```python
sns.countplot(tips['smoker'])      # used to count the frequency of a particular parameter
```

```python
sns.countplot(tips['smoker'],data=tips,hue='sex')
```

## 3.Boxplot

```python
sns.boxplot('total_bill',data=tips)
```

```python
sns.boxplot('total_bill',data=tips,hue='sex')
```

```python
sns.boxplot(x='total_bill',y='sex',data=tips)
```

```python
sns.boxplot(x='sex',y='tip',data=tips)
```

## 4.Violinplot

```python
sns.violinplot(x='sex',y='tip',data=tips)
```

```python
sns.violinplot(x='day',y='tip',data=tips)
```

```python
sns.violinplot(x='day',y='tip',data=tips)
```

# MATRIX PLOT

## 1. Heatmap

```python
flight=sns.load_dataset('flights')
```

```python
flight.head()
```

```python
import pandas as pd
```

```python
p=flight.pivot_table(index='year',columns='month',values='passengers')
```

```python
p
```

```python
sns.heatmap(p)
```

```python
sns.heatmap(p,annot=True)
```

```python
sns.heatmap(p, linewidths=3, linecolor= 'black', cmap= 'summer')
```

```python
sns.heatmap(p, linewidths=3, cmap= 'autumn')
```
