# Chances of Female survived

```python
survival=titanic.groupby('Sex').sum()['Survived']
survival
```

```python
total=titanic['Sex'].value_counts()
total
```

```python
# probability of male and female survival
```

```python
x=survival/total
```

```python
print (x)
```

```python
x.plot(kind='bar',title='Chance of survival')
```
